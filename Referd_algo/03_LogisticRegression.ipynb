```python
combined_df = pd.read_pickle('combined.pkl')
unified_combined_df = pd.read_pickle('unified_combined.pkl')
seg_scan_swipe_df = pd.read_pickle('seg_scan_swipe.pkl')
unified_col_seg_df = pd.read_pickle('unified_col_seg.pkl')

dfs = [combined_df, unified_combined_df, seg_scan_swipe_df, unified_col_seg_df]
for df in dfs:
    pca = PCA(n_components=0.9)
    df_pca_transformed = pd.DataFrame(pca.fit_transform(df.drop(columns=['like'])))
    df_pca_transformed['like'] = df.like

    df_like_data = df_pca_transformed[~df_pca_transformed.like.isna()]
    X_train, X_test, y_train, y_test = train_test_split(
        df_like_data.drop(columns=['like']), df_like_data.like, test_size=0.3
    )
    clf = LogisticRegression()
    clf.fit(X_train, y_train)
    print('ACCURACY')
    print(accuracy_score(clf.predict(X_train), y_train))
    print(accuracy_score(clf.predict(X_test), y_test))
    print('PRECISION')
    print(precision_score(clf.predict(X_train), y_train))
    print(precision_score(clf.predict(X_train), y_train))
    print('RECALL')
    print(recall_score(clf.predict(X_train), y_train))
    print(recall_score(clf.predict(X_test), y_test))
    print('PROPORTION OF PREDICTED RIGHT SWIPES')
    df_no_like_data_X = df_pca_transformed[df_pca_transformed.like.isna()].drop(columns=['like'])
    print(len(list(filter(lambda i: not i, clf.predict(df_no_like_data_X))))/df_no_like_data_X.shape[0])
```
